```python
import numpy as np
import math
import pandas as pd
import matplotlib.pyplot as mp
import statsmodels.api as sm
import seaborn
seaborn.set()
```

```python
raw_data=pd.read_csv('fifa19.csv',nrows=50)
raw_data#Contains all data
```

```python
y=raw_data['Wage']
x1=raw_data[['Overall','Age']]
#x1=np.log10(x1)
x1=np.log10(x1)
y=np.log10(y)
#y=np.log10(y)
mp.scatter(y,x1['Overall'])
mp.show()
```

```python
#Lets apply linear regression to see how apt the result can be
x=sm.add_constant(x1)
#x=float(math.log10(x))
result=sm.OLS(y,x).fit()
result.summary()
```

```python
new_data=pd.DataFrame({'const':1, 'Overall':[94,94], 'Age':[33,31]})
new_data
```

```python
prediction=result.predict(new_data)
prediction
```
